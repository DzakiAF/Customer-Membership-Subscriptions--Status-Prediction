## Inference with actual data from dataset

```python
dumm = pd.read_csv("shopping_behavior_updated.csv")
```

```python
dumm['Subscription Status'].unique()
```

```python
dumm.drop(['Customer ID'], axis=1, inplace=True)
dumm.drop(['Location'], axis=1, inplace=True)
dumm = dumm.rename(columns={'Purchase Amount (USD)': 'Purchase Amount'})
```

```python
subs_yes = dumm[dumm['Subscription Status']== 'Yes']
```

```python
subs_no = dumm[dumm['Subscription Status']== 'No']
```

### Predict with original data data where Subs Status == Yes.

```python
y_pred = model3.predict(subs_yes)
np.set_printoptions(threshold=np.inf)
y_pred
```

### Predict with original data where Subs Status == No.

```python
y_pred = model3.predict(subs_no)
np.set_printoptions(threshold=np.inf)
y_pred
```
